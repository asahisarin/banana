```python
import pandas as pd

df = pd.read_csv("/content/drive/My Drive/banana/banana_quality.csv")
df.head(5)
```

```python
df.drop('Sweetness', axis=1, inplace=True)
```

```python
df.head()
```

```python
X = df[['Size','Weight','Softness']]
y = df['Quality']

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test=train_test_split(X, y, test_size=0.1, random_state=123)
```

```python
from sklearn.tree import DecisionTreeClassifier

#membuat model decision tree
tree_model= DecisionTreeClassifier()

#melatih model menggunakan data latih dengan fungsi fit
tree_model= tree_model.fit(X_train,y_train)
```

```python
#evaluasi model untuk meguji seberapa baik model yang digunakan dan akurasi

from sklearn.metrics import accuracy_score
y_pred =  tree_model.predict(X_test)
accuracy_score =round(accuracy_score(y_pred, y_test), 3)
print('Accuracy: ',accuracy_score)
```

```python
#prediksi model dengan tree model yang sdh dibuat dengan varibel (['Size','Weight','Softness'])

print(tree_model.predict([[-1.924968, 0.468078, -1.472177]])[0])
```

```python
print(tree_model.predict([[-0.468078, -2.495099, -2.645145]])[0])
```

```python
#melihat visulisasi datanya decision tree
from sklearn.tree import export_graphviz

export_graphviz(tree_model,
                out_file="df_tree.dot",
                feature_names=['Size','Weight','Softness'],
                class_names=['good','bad'],
                rounded=True,
                filled=True)
```
